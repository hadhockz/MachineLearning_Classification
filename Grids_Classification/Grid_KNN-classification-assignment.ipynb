```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
#printing the dataset
dataset
```

```python
#converting the classification categorical value to numerical value.
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#printing the updated dataset
dataset
```

```python
#renaming the columns
dataset.rename(columns={"rbc_normal":"rbc","pc_normal":"pc","pcc_present":"pcc","ba_present":"ba","htn_yes":"htn",
                        "dm_yes":"dm","cad_yes":"cad","appet_yes":"appet","pe_yes":"pe",
                        "ane_yes":"ane","classification_yes":"classification"},inplace=True)
```

```python
#printing the updated dataset
dataset
```

```python
#printing the columns
dataset.columns
```

```python
#printing the classification to know whether it is balanced or imbalanced.
dataset["classification"].value_counts()
```

```python
#separating the feature and target values.
feature = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc', 'pc', 'pcc', 'ba',
       'htn', 'dm', 'cad', 'appet', 'pe', 'ane']]
target = dataset[['classification']]
```

```python
#printing the no of columns and rows of input and output 
feature.shape,target.shape
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size = 0.33, random_state = 42)
```

```python
(X_train,X_test),(X_train.shape,X_test.shape)
```

```python
(Y_train,Y_test),(Y_train.shape,Y_test.shape)
```

```python
#model creation and learning phase
from sklearn.neighbors import KNeighborsClassifier
#importing the gridsearch to fit hyptun.parameter in table and 
#find best parameter
from sklearn.model_selection import GridSearchCV
hyperparam_grid = {'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
                'weights' : ['uniform','distance']}         
grid_KNN = GridSearchCV(KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p = 2),
                        hyperparam_grid,cv=3, refit=True,verbose=3, n_jobs=-1,scoring='f1_weighted')

#Model creation grid search
grid_KNN.fit(X_train, Y_train)
        
```

```python
#printing the best parameter after tuning.
re=grid_KNN.cv_results_
Y_prediction = grid_KNN.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_prediction)

#printing the classificaton report
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, Y_prediction)
```

```python
from sklearn.metrics import f1_score
macro_value = f1_score(Y_test,Y_prediction,average='weighted')
print("The macro_value for best parameter {}:".format(grid_KNN.best_params_),macro_value)
```
