```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the CSV file
dataset = pd.read_csv("CKD.csv")
```

```python
#printing the fetched dataset
dataset
```

```python
#converting categorical to numerical value
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
#printing the updated dataset
dataset
```

```python
#renaming the columns
dataset.rename(columns={"rbc_normal":"rbc","pc_normal":"pc","pcc_present":"pcc","ba_present":"ba","htn_yes":"htn",
                        "dm_yes":"dm","cad_yes":"cad","appet_yes":"appet","pe_yes":"pe",
                        "ane_yes":"ane","classification_yes":"classification"},inplace=True)
```

```python
#printing the updated columns
dataset
```

```python
#printing the columns
dataset.columns
```

```python
#separating the feature and target values
feature = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc', 'pc', 'pcc', 'ba',
       'htn', 'dm', 'cad', 'appet', 'pe', 'ane']]
target = dataset[['classification']]
```

```python
#printing the shape(rows,columns) feature and target values
feature.shape, target.shape
```

```python
#printing the feature and target values
print(feature,target)
```

```python
#splitting the train and test set to train the model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(feature,target,test_size=0.33,random_state=42)
```

```python
#printing the train and test set
X_train,X_test
```

```python
Y_train,Y_test
```

```python
#model creation and learning phase
from sklearn.ensemble import RandomForestClassifier
#importing gridsearch to fit hypertunning parameter's in table form automatically &
#To find best parameter. 
from sklearn.model_selection import GridSearchCV

hyperparam_grid = {'criterion': ['gini','entropy'],
                  'max_features':['auto','sqrt','log2'],
                  'n_estimators':[10,100]}

grid_rf = GridSearchCV(RandomForestClassifier(),
                        hyperparam_grid,cv=3, refit=True,verbose=3,
                        n_jobs=-1,scoring='f1_weighted')

#Model creation for grid search
grid_rf.fit(X_train,Y_train)
```

```python
#printing the best parameter after tuning
re=grid_rf.cv_results_
Y_prediction = grid_rf.predict(X_test)

from sklearn.metrics import confusion_matrix
cm= confusion_matrix(Y_test,Y_prediction)

#printing classification report
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test,Y_prediction)
```

```python
from sklearn.metrics import f1_score
macro_value = f1_score(Y_test,Y_prediction,average='weighted')
print("The macro_value for best parameter {}:".format(grid_rf.best_params_),macro_value)
```
