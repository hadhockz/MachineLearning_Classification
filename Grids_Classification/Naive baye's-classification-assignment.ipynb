```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#renaming the columns
dataset.rename(columns={"rbc_normal":"rbc","pc_normal":"pc","pcc_present":"pcc","ba_present":"ba","htn_yes":"htn",
                        "dm_yes":"dm","cad_yes":"cad","appet_yes":"appet","pe_yes":"pe",
                        "ane_yes":"ane","classification_yes":"classification"},inplace=True)
```

```python
#printing the updated dataset
dataset
```

```python
#printing the columns
dataset.columns
```

```python
dataset["classification"].value_counts()
```

```python
#separating the feature and target values.
feature = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc', 'pc', 'pcc', 'ba',
       'htn', 'dm', 'cad', 'appet', 'pe', 'ane']]
target = dataset[['classification']]
```

```python
#printing the no of columns and rows of input and output 
feature.shape,target.shape
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size = 0.33, random_state = 42)
```

```python
(X_train,X_test),(X_train.shape,X_test.shape)
```

```python
(Y_train,Y_test),(Y_train.shape,Y_test.shape)
```

```python
#Model creation & Training phase
from sklearn.naive_bayes import MultinomialNB
MNB_classif = MultinomialNB()
MNB_classif.fit(X_train, Y_train)

#Testing phase
Y_pred = MNB_classif.predict(X_test)


#Scoring phase for decide best model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, Y_pred)              

print(clf_report)

print(cm)
```

```python
#support values [53 79] denotes the Y_test which is 132(rows) x 1(cols) - Accuracy is 93 for MNB
```

```python
#Model creation & Training phase
from sklearn.naive_bayes import BernoulliNB
BNB_classif = BernoulliNB()
BNB_classif.fit(X_train, Y_train)

#Testing phase
Y_pred = BNB_classif.predict(X_test)


#Scoring phase for decide best model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, Y_pred)    

print(clf_report)

print(cm)
```

```python
#support values [53 79] denotes the Y_test which is 132(rows) x 1(cols) - Accuracy is .91 for BNB
```

```python
#MOdel creation and training phase
from sklearn.naive_bayes import ComplementNB
CMNB_classif =ComplementNB()
CMNB_classif .fit(X_train, Y_train)

#Testing phase
Y_pred = CMNB_classif.predict(X_test)

#SCoring phase to choose best model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, Y_pred)    

print(clf_report)

print(cm)
```
